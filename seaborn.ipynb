```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set(style="darkgrid")
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.relplot(x="total_bill", y="tip", data = tips)
```

```python
tips["smoker"].value_counts()
```

if we want to change the colour of the dots in to the graph

```python
sns.relplot(x="total_bill", y= "tip", data= tips, hue="smoker")
```

if we want to change the style of the dots in to the graph

```python
sns.relplot(x="total_bill", y= "tip", data= tips, style="time")
```

```python
sns.relplot(x="total_bill", y="tip", hue="size", data= tips)
```

if we want to change the colour of the size dots in to the graph

```python
sns.relplot(x="total_bill", y="tip", hue="size", data= tips, palette="ch:r=0.9, l=0.9")
```

if we want to change the size of the dots into the graph 

```python
sns.relplot(x="total_bill", y="tip", size="size", data= tips)
```

```python
sns.relplot(x="total_bill", y="tip", data= tips, size="size", sizes=(15,200))
```

```python
from numpy.random import randn
```

```python
df= pd.DataFrame(dict(time= np.arange(500), value= randn(500).cumsum()))
```

```python
df.head()
```

here we are making line plot 

```python
sns.relplot(x="time", y="value", kind= "line", data=df)
```

we are making the new dataframe to check the sort filter in graph, here time is not sorted 

```python
df=pd.DataFrame(randn(500,2).cumsum(axis=0), columns=["time","value"])
```

```python
df.head()
```

```python
sns.relplot(x="time", y="value", kind= "line", data=df, sort=False)
```

```python
fmri= sns.load_dataset("fmri")
fmri.head()
```

here we are ploating the data of multiple value in same time

```python
sns.relplot(x="timepoint", y="signal", kind="line", data=fmri)
```

here we are plot a graph with confidence intervel

```python
sns.relplot(x="timepoint", y="signal", kind="line", data=fmri, ci="sd")
```

graph without confidence interval

```python
sns.relplot(x="timepoint", y="signal", estimator=None, kind="line", data=fmri)
```

```python
sns.relplot(x="timepoint", y="signal", hue="event" , kind="line", data=fmri)
```

```python
sns.relplot(x="timepoint", y="signal", hue="region", style="event" , kind="line", data=fmri)
```

```python
sns.relplot(x="timepoint", y="signal", hue="region", style="event" , kind="line", data=fmri, markers=True)
```

```python
dots=sns.load_dataset('dots').query("align == 'dots'")
```

```python
dots.head()
```

```python
sns.relplot(x="time", y= "firing_rate", hue= "coherence", data= dots, kind= "line", style= "choice")
```

if we want to change the colour of graph lines

```python
palette= sns.cubehelix_palette(light=0.5, n_colors=6)
sns.relplot(x="time", y= "firing_rate", hue= "coherence", data= dots, kind= "line", style= "choice", palette=palette)
```
